```python
#import libraries
import PIL
from PIL import Image
import matplotlib.pyplot as plt

#function to plot histogram
def histogram(img):
  pixels=[]

  #create list of values 0-255
  for x in range(256):
    pixels.append(x)
  
  #initialize width and height of image
  width,height=img.size
  counts=[]
  
  #for each intensity value
  for i in pixels:
    #set counter to 0
    temp=0

    #traverse through the pixels
    for x in range(width):
      for y in range(height):

        #if pixel intensity equal to intensity level
        #increment counter
        if (img.getpixel((x,y))==i):
          temp=temp+1

    #append frequency of intensity level 
    counts.append(temp)
  
  #plot histogram
  plt.bar(pixels,counts)
  plt.show()
```

```python
import imageio.v3 as iio
import numpy as np
import skimage.color
import skimage.util
import matplotlib.pyplot as plt


# read the image of a plant seedling as grayscale from the outset
image = iio.imread(uri="/content/drive/MyDrive/diphw1/sp.jpg", mode="L")

# convert the image to float dtype with a value range from 0 to 1
image = skimage.util.img_as_float(image)

# display the image
fig, ax = plt.subplots()
plt.imshow(image, cmap="gray")
```

In an image processing context, the histogram of an image normally refers to a histogram of the pixel intensity values. This histogram is a graph showing the number of pixels in an image at each different intensity value found in that image.

```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
hist, bins = np.histogram(x, bins=50)
width = 0.7 * (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, hist, align='center', width=width)
plt.show()
```

```python
import imageio  as iio
import skimage.color
import skimage.filters


# load the image
load_image = iio.imread(uri="/content/drive/MyDrive/diphw1/lena.jpg")

fig1, ax1 = pl.subplots()
pl.imshow(load_image)
```

```python
# image to grayscale conversion
grayScale_image = skimage.color.rgb2gray(load_image)

# picture blurring to remove noise
blur_img = skimage.filters.gaussian(grayScale_image, sigma=.5)

fig, ax = pl.subplots()
pl.imshow(blur_img, cmap="gray")
```

```python
histogram1, bin_edges = npy.histogram(blur_img, bins=256, range=(0.0, 1.0))

fig1, ax1 = pl.subplots()
pl.plot(bin_edges[0:-1], histogram1)
pl.title("Grayscale Histogram")
pl.xlabel("grayscale value")
pl.ylabel("pixels")
pl.xlim(0, 1.0)
```

```python
for i in range(1,8):
  threshold= i/10
  binary_mask_threshold = blur_img < threshold
  fig1, ax1 = pl.subplots()
  I= pl.imshow(binary_mask_threshold, cmap="gray")
```

```python
histogram2, bin_edges = npy.histogram(binary_mask_threshold, bins=256, range=(0.0, 1.0))
#histogram2, bin_edges = npy.histogram(binary_mask_threshold, bins=256)
fig2, ax2 = pl.subplots()
pl.plot(bin_edges[0:-1], histogram2)
pl.title("Grayscale Histogram")
pl.xlabel("grayscale value")
pl.ylabel("pixels")
pl.xlim(0, 1.0)
```

```python
for i in range(1,8):
  threshold= (8-i-1)/10
  binary_mask_threshold_1 = blur_img < threshold
  fig3, ax3 = pl.subplots()
  J= pl.imshow(binary_mask_threshold_1, cmap="gray")
```
