#Problem 2: Harris Corner Detection

```python
import numpy
import matplotlib.pyplot
import cv2
```

```python
from PIL.Image import *
from scipy import ndimage
```

```python
def Image_map(I):
     x= Image.getpixel(I, (0, 0))
     if type(x) == int:
        return I
     else:
        c, l = I.size
        I_Array = numpy.asarray(I)
        neim = numpy.zeros((l, c))
        for i in range(l):
            for j in range(c):
                ti = I_Array[i, j]
                tsec = sum(ti)/len(ti)
                neim[i, j] = tsec
        return neim
```

```python
def H_C_D(I):
    neim = Image_map(I)
    I_Array = numpy.asarray(neim, dtype=numpy.float64)
    Ix = ndimage.sobel(I_Array, 0)
    Iy = ndimage.sobel(I_Array, 1)
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy
    Ix2 = ndimage.gaussian_filter(Ix2, sigma=2)
    Iy2 = ndimage.gaussian_filter(Iy2, sigma=2)
    Ixy = ndimage.gaussian_filter(Ixy, sigma=2)
    c, l = I_Array.shape
    Res = numpy.zeros((c, l))
    R = numpy.zeros((c, l))
    R_max = 0
    for i in range(c):
        for j in range(l):
            print('test ',j)
            Mat = numpy.array([[Ix2[i, j], Ixy[i, j]], [Ixy[i, j], Iy2[i, j]]], dtype=numpy.float64)
            R[i, j] = numpy.linalg.det(Mat) - 0.04 * (numpy.power(numpy.trace(Mat), 2))
            if R[i, j] > R_max:
                R_max = R[i, j]
    for i in range(c - 1):
        for j in range(l - 1):
          if R[i, j] > 0.01 * R_max and R[i, j] > R[i-1, j-1] and R[i, j] > R[i-1, j+1]\
                                    and R[i, j] > R[i+1, j-1] and R[i, j] > R[i+1, j+1]:
                                    Res[i, j] = 1

    PC, PR = numpy.where(Res == 1)
    matplotlib.pyplot.plot(PR, PC, 'r+')
    matplotlib.pyplot.imshow(I, 'gray')
    matplotlib.pyplot.show()
```
