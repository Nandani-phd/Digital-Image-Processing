Problem 3.1:Digital Video

```python
import cv2
```

```python
capture_img = cv2.VideoCapture('/content/drive/Mydrive/diphw/vid1.mp4')
```

```python
if capture_img.isOpened():
    cf = 0
    while True:
        ret, frame = capture_img.read()
        if ret:
            name = f'frameIn/frame{cf}.jpg'
            print(f"Creating file... {name}")
            cv2.imwrite(name, frame)
            frame.append(name)
        cf += 1
    capture_img.release()
cv2.destroyAllWindows()
```
