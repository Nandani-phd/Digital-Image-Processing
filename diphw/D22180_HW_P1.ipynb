```python
#Problem 1.5: Gray Image
```

```python
import cv2
```

```python
import numpy as npy
```

```python
img5 = cv2.imread('/content/drive/MyDrive/diphw/lena.jpg')
I = img5.reshape((-1,3))
```

```python
I = npy.float32(I)
import matplotlib.pyplot as pl
```

```python
#ctr = criteria, c= cluster, l=level, cen=center
ctr = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
def colorQuant(I, c, ctr):
  ret,l,cen =cv2.kmeans(I,c,None,ctr,10,cv2.KMEANS_RANDOM_CENTERS)
  cen = npy.uint8(cen)
  res = cen[l.flatten()]
  res1 = res.reshape((img5.shape))
  return res1
res0 = colorQuant(I, 1, ctr)
res1 = colorQuant(I, 2, ctr)
res2 = colorQuant(I, 3, ctr)

pl.subplot(221),pl.imshow(cv2.cvtColor(img5, cv2.COLOR_BGR2RGB))
pl.title('Original Image'), pl.xticks([]), pl.yticks([])

pl.subplot(222),pl.imshow(cv2.cvtColor(res0, cv2.COLOR_BGR2RGB))
pl.title('K=2'), pl.xticks([]), pl.yticks([])

pl.subplot(223),pl.imshow(cv2.cvtColor(res1, cv2.COLOR_BGR2RGB))
pl.title('K=4'), pl.xticks([]), pl.yticks([])

pl.subplot(224),pl.imshow(cv2.cvtColor(res2, cv2.COLOR_BGR2RGB))
pl.title('K=8'), pl.xticks([]), pl.yticks([])
pl.show()
```
